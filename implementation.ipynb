```python
X=df.drop(columns=['Delivery_Time_min','Order_ID'])
```

```python
y=df[['Delivery_Time_min']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(df[['Weather','Traffic_Level','Time_of_Day','Vehicle_Type']])
ct=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['Weather','Traffic_Level','Time_of_Day','Vehicle_Type']),remainder='passthrough')

model=LinearRegression()
pipe=make_pipeline(ct,model)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
scores=[]
for i in range(1,101):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    y_pred=pd.DataFrame(data=y_pred,columns=['prediction'])
    result=pd.concat([y_test.reset_index(drop=True),y_pred],axis=1)
    score=r2_score(result['Delivery_Time_min'],result['prediction'])
    scores.append(score)
    print(i,score)
```

```python
print(max(scores))
```
